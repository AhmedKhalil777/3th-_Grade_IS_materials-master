```python
import os 

def mdarr():
    f= open('mcq.txt','r')
    if f.mode == 'r':
        contents = f.read()
        conarr = contents.rsplit('\n')
        scon = divide_chunks(conarr,8)
        mcq =''
        for x in scon :
            x[0] = '## '+x[0]+'\r\n'
            d = x[6].split(')')
            x[6] ='![](<label style="color:green; font :bold 30px;"> '+d[0]+d[1]+'</label>)'+'\r\n'
            x[7] ='\r\n---\r\n'
            for i in range(1,6):
                x[i] = ' - ' + x[i]+'\r\n'
            for i in x:
                mcq +=i;
        mcqf = open('mcq.md', 'w+')
        mcqf.write(mcq)
def divide_chunks(l, n): 
      
    # looping till length l 
    for i in range(0, len(l), n):  
        yield l[i:i + n]     

if __name__ == "__main__":
    mdarr()
    input("finish")
```
